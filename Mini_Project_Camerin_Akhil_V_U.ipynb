# **Laptop Price Prediction**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

from sklearn.model_selection import train_test_split,KFold,StratifiedKFold,cross_val_score,GridSearchCV
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
df=pd.read_csv('/content/drive/MyDrive/Camerin DSML/laptops.csv')
df.head()
```

**Preprocessing**

```python
df=df.drop(columns=['img_link','Unnamed: 0'])
df.head()
```

```python
df['name']=df['name'].str.upper()
df.head()
```

```python
df['processor']=df['processor'].str.upper()
df.head()
```

```python
df['os']=df['os'].str.upper()
df.head()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df['rating']=df['rating'].fillna((df['rating'].mean()))
df['no_of_ratings']=df['no_of_ratings'].fillna((df['no_of_ratings'].mean()))
df['no_of_reviews']=df['no_of_reviews'].fillna((df['no_of_reviews'].mean()))
```

```python
df1=df
df1.head()
```

```python
temp=df['price(in Rs.)']
df=df.drop(columns='price(in Rs.)')
```

```python
label_encoder=preprocessing.LabelEncoder()
for i in df:
  df[i]=label_encoder.fit_transform(df[i])
```

```python
df.dtypes
```

```python
df.head()
```

```python
scaler=StandardScaler()
for i in df:
  df[[i]]=scaler.fit_transform(df[[i]])
df.head()
```

```python
df['price(in Rs.)']=temp
df.head()
```

```python
x=df.drop(columns='price(in Rs.)')
x.head()
```

```python
y=df['price(in Rs.)']
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=7)
```

**Model Selection**

```python
LR_score=cross_val_score(LinearRegression(),X=x,y=y,cv=5)
DT_score=cross_val_score(DecisionTreeRegressor(),X=x,y=y,cv=5)
RF_score=cross_val_score(RandomForestRegressor(),X=x,y=y,cv=5)
KN_score=cross_val_score(KNeighborsRegressor(),X=x,y=y,cv=5)
AD_score=cross_val_score(AdaBoostRegressor(),X=x,y=y,cv=5)
GB_score=cross_val_score(GradientBoostingRegressor(),X=x,y=y,cv=5)
XG_score=cross_val_score(XGBRegressor(),X=x,y=y,cv=5)
svm_score=cross_val_score(SVR(),X=x,y=y,cv=5)
```

```python
print(f'Linear Regression : {np.round(np.mean(LR_score)*100,2)}%')
print(f'Decision Tree : {np.round(np.mean(DT_score)*100,2)}%')
print(f'Random Forest : {np.round(np.mean(RF_score)*100,2)}%')
print(f'KNN : {np.round(np.mean(KN_score)*100,2)}%')
print(f'Adaboost : {np.round(np.mean(AD_score)*100,2)}%')
print(f'Gradient Boost : {np.round(np.mean(GB_score)*100,2)}%')
print(f'XGB : {np.round(np.mean(XG_score)*100,2)}%')
print(f'SVM : {np.round(np.mean(svm_score)*100,2)}%')
```

We select 3 models:
* Random Forest : 78.88%
* Gradient Boost : 80.64%
* XGB : 81.43%

**Model Evaluation**

```python
rf=RandomForestRegressor()
forest_params=[{'max_depth':list(range(10,15)),'max_features':list(range(0,15))}]
```

```python
clf=GridSearchCV(rf,forest_params,cv=5,scoring='r2')
clf.fit(x_train,y_train)
print(clf.best_params_)
print(clf.best_score_)
```

{'max_depth': 12, 'max_features': 2}
0.8212935129263375

```python
gb=GradientBoostingRegressor()
gb_params=[{'max_depth':list(range(2,7)),'learning_rate':[i/100 for i in range(1,21)]}] #(0.01-0.2)
```

```python
clf=GridSearchCV(gb,gb_params,cv=5,scoring='r2')
clf.fit(x_train,y_train)
print(clf.best_params_)
print(clf.best_score_)
```

{'learning_rate': 0.15, 'max_depth': 3}
0.834320915972992

```python
xg=XGBRegressor()
xg_params=[{'colsample_bytree':[i/10 for i in range(4,11)],'gamma':list(range(0,10))}]
```

```python
clf=GridSearchCV(xg,xg_params,scoring='r2')
clf.fit(x_train,y_train)
print(clf.best_params_)
print(clf.best_score_)
```

{'colsample_bytree': 0.7, 'gamma': 0}
0.806666910648346

```python
GB=GradientBoostingRegressor(learning_rate=0.15,max_depth=3)
GB.fit(x_train,y_train)

GB_pred=GB.predict(x_test)
GB_pred
```

```python
a=GB.score(x_test,y_test)
print(f'Accuracy of the Gradient Boost Algorithm : {np.round((a*100),2)}%')
```

```python
RF=RandomForestRegressor(max_depth=14,max_features=3)
RF.fit(x_train,y_train)

RF_pred=RF.predict(x_test)
RF_pred
```

```python
a=RF.score(x_test,y_test)
print(f'Random Forest Algorithm : {np.round((a*100),2)}%')
```
